```python
ticker = yf.Ticker('BBAS3.sa')
data = ticker.history(start='2010-01-01', end='2021-09-01')
```

```python
data['Close'].plot()
```

```python
#variação da coração da empresa de 2016 - 2021
(data['Close'][-1] - data['Close'][0]) / data['Close'][0] * 100
```

```python
#dividendos são proventos pagos pela empresa aos seus acionistas com base no lucro aferido
#dividendos e JCP são tipos de proventos
dataDiv = data['Dividends'].resample('3M').sum()
dataDiv.plot()
```

```python
dataCot = data['Close'].resample('3M').median()
```

```python
dataCot.plot()
```

```python
#dataframe é um tipo de objeto do pandas para organizar dados
#DataFrame() é um método para criar dataframes a partir de series, listas, dict, etc
dataCot = pd.DataFrame(dataCot)
dataDiv = pd.DataFrame(dataDiv)
```

```python
#concat é um método do pandas para juntar dataframes
dataJoin = pd.concat([dataCot, dataDiv], axis=1)
```

```python
dataJoin['Close'] = np.log10(dataJoin['Close'])
dataJoin['Dividends'] = np.log10(dataJoin['Dividends'])

dataJoin.plot()
```

```python
#utilize o minMaxScaller do sklearn para comparar a variação dividendosXcotação

dataJoin2 = dataJoin

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

dataJoin2 = dataJoin2[np.isfinite(dataJoin2).all(1)]

dataJoin2[['Close', 'Dividends']] = scaler.fit_transform(dataJoin2[['Close', 'Dividends']])

#dataJoin2

dataJoin2.plot()
```
