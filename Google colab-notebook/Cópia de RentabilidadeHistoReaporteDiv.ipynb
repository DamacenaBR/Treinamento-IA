```python
def reaporteDivPlusValue(ticket, value):
    
    comp = yf.Ticker(ticket)
    #hist = comp.history(period="10y")
    hist = comp.history(start='2011-01-01', end='2015-12-30')
    
    if(len(hist) > 0 and hist.index[0].year <= 2011 and hist.index[0].month == 1):
        somaDiv = hist['Dividends'].resample('M').sum()
        meanPrice = hist['Close'].resample('M').median()
    
        saldoDin = value 
        saldoQtde = saldoDin // meanPrice[0]
        caixa = saldoDin - (saldoQtde * meanPrice[0])
    
        for i in range(1, len(somaDiv)):
            caixa = caixa + value + (somaDiv[i] * saldoQtde)
            if(caixa > meanPrice[i]):
                novaQtde = (caixa // meanPrice[i]) #quantidade de ações novas que eu comprei naquele mês
                saldoQtde = saldoQtde + novaQtde # somar a nova qtde com o que eu já possuia
                caixa = caixa - (novaQtde * meanPrice[i]) # o que restou de dinheiro após comprar as ações naquele mês
                patr = caixa + saldoQtde * meanPrice[i]
                print('qtdeAcoes:{:.2f} -caixa:{:.2f} -patrim:{:.2f}'.format(saldoQtde, caixa, patr))

        patr = caixa + saldoQtde * meanPrice[len(meanPrice)-1]
        
        #print('{} = PatrFinal: {:.2f} - rent:{:.2f}%'.format(ticket, patr, (patr - saldoDin)/saldoDin * 100)) 
        return patr
    
    return -999.999
```

```python
dados = pd.read_csv('/content/statusinvest09102021.csv', decimal="," , delimiter=";", thousands=".")
dados = dados.fillna(0) 
```

```python
dados.drop(dados[dados[' LIQUIDEZ MEDIA DIARIA'] < 1000000].index, inplace=True)
dados.drop(dados[dados['DY'] < 3].index, inplace=True)
dados.drop(dados[dados['PRECO'] == 0].index, inplace=True)
```

```python
dataf = []
for empresa in dados['TICKER']:
    dataf.append([empresa, reaporteDivPlusValue(empresa + '.sa', 100)])
```

```python
retornos = pd.DataFrame(dataf, columns=['ticker','rentb'])
retornos.drop(retornos[retornos['rentb'] < -900].index, inplace=True)
```
